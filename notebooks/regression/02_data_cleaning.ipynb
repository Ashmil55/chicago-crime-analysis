# Regression - Data Cleaning

Prepare the Chicago crime data for regression-style modeling by cleaning missing values, creating time features, and saving a processed dataset.

```python
from pathlib import Path
import sys

import pandas as pd

PROJECT_ROOT = Path.cwd().resolve().parents[1] if Path.cwd().name == 'regression' else Path.cwd().resolve()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from src.load_data import load_raw_crimes
from src.preprocess import clean_data, save_cleaned

RAW_PATH = PROJECT_ROOT / 'data' / 'raw' / 'crimes.csv'
PROCESSED_PATH = PROJECT_ROOT / 'data' / 'processed' / 'crimes_cleaned.csv'
```

```python
df_raw = load_raw_crimes(RAW_PATH)
print('Raw shape:', df_raw.shape)
df_raw.head()
```

```python
df_clean = clean_data(df_raw)
print('Cleaned shape:', df_clean.shape)
df_clean.head()
```

```python
comparison = pd.DataFrame({
    'raw_missing': df_raw.isna().sum(),
    'clean_missing': df_clean.isna().sum()
}).fillna(0).astype(int)
comparison.sort_values('raw_missing', ascending=False).head(20)
```

```python
important_columns = [
    'Date', 'Primary Type', 'Beat', 'Community Area', 'Latitude', 'Longitude',
    'Hour', 'DayOfWeek', 'Month', 'WeekOfYear', 'IsWeekend', 'HourBin', 'Crime_Category'
]
df_clean[important_columns].describe(include='all').T
```

```python
saved_path = save_cleaned(df_clean, PROCESSED_PATH)
saved_path
```
