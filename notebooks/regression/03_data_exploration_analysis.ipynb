# Regression - Data Exploration Analysis

Explore spatial, categorical, and temporal behavior in the cleaned crime dataset to inform feature engineering for regression and prediction tasks.

```python
from pathlib import Path
import sys

import matplotlib.pyplot as plt
import pandas as pd

PROJECT_ROOT = Path.cwd().resolve().parents[1] if Path.cwd().name == 'regression' else Path.cwd().resolve()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from src.eda import (
    load_for_eda,
    plot_arrest_domestic,
    plot_crime_by_day_of_week,
    plot_crime_by_hour,
    plot_crime_density_hexbin,
    plot_location_types,
    plot_primary_type_distribution,
    plot_top_beats,
)

CLEANED_PATH = PROJECT_ROOT / 'data' / 'processed' / 'crimes_cleaned.csv'
```

```python
df = load_for_eda(CLEANED_PATH)
print('EDA shape:', df.shape)
df.head()
```

```python
monthly_volume = df.groupby(['Year', 'Month']).size().reset_index(name='crime_count')
monthly_volume.head()
```

```python
plt.figure(figsize=(14, 5))
monthly_volume['period'] = monthly_volume['Year'].astype(str) + '-' + monthly_volume['Month'].astype(str).str.zfill(2)
plt.plot(monthly_volume['period'], monthly_volume['crime_count'])
plt.title('Monthly Crime Volume')
plt.xlabel('Period')
plt.ylabel('Crime Count')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plot_primary_type_distribution(df)
plot_location_types(df)
plot_top_beats(df)
plot_crime_by_hour(df)
plot_crime_by_day_of_week(df)
plot_arrest_domestic(df)
plot_crime_density_hexbin(df)
```

```python
regression_aggregates = {
    'beat_day_counts': df.groupby(['Beat', 'DayOfWeek']).size().sort_values(ascending=False).head(10),
    'community_month_counts': df.groupby(['Community Area', 'Month']).size().sort_values(ascending=False).head(10),
    'hour_location_counts': df.groupby(['Hour', 'Location Description']).size().sort_values(ascending=False).head(10),
}

regression_aggregates
```
